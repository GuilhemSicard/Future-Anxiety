```python
import os, json
import pandas as pd
import json 
import plotly.express as px
from IPython.display import Image
import re
import gc
```

```python
def filterPage(pages):
    if len(pages) == 4 :
        return [(page,content) for page,content in enumerate(pages[:-1])]
    elif (len(pages) == 6) or (len(pages) == 8):
        return [(page,content) for page,content in enumerate(pages[:-2])]
    else : 
        return []
    
def getOCRconfidence(confidence):
    return float(confidence[0].split()[-1].replace(',','.'))

def getAlphaRatio(content):
    return sum([c.isalpha() and c.islower() for c in content])/len(content)

def removeNoisySection(text):
    textBlocks =((getAlphaRatio(textBlock),textBlock) 
                 for textBlock 
                 in text.split('\n') 
                 if len(textBlock)>0)
    textBlocks = (textBlock 
                  for alphaRatio, textBlock 
                  in textBlocks
                  if ((alphaRatio>0.7) and (len(textBlock)<=200)) or ((alphaRatio>0.5) and (len(textBlock)>200)))
    return '\n'.join(textBlocks)
```

```python
text = []
for root, dirs, files in os.walk('data/'):
    for name in files:
        if name.endswith((".json")) and 'le_figaro' not in name:
            with open(os.path.join(root, name)) as f: 
                text.append(json.load(f))
```

```python
text = pd.DataFrame(text)
gc.collect()
text['date'] = text.date.apply(lambda x : x[0]) 
text['year'] = text.date.apply(lambda x : x.split('-')[0]) 
text['OCR confidence'] = text['format'].apply(getOCRconfidence)
text['nPages'] = text.contentAsText.apply(len)
text = text[['contentAsText','date','OCR confidence','nPages']]
gc.collect()
```

```python
img = px.scatter(text, x = 'date', y = 'nPages',opacity = 1)
#img.update_traces(width=100)
Image(img.to_image(format="png", width=600, height=350, scale=2))
```

```python
textDf = text.drop(columns = 'nPages').copy()
textDf.contentAsText = textDf.contentAsText.apply(filterPage)
textDf = textDf[textDf.contentAsText.apply(len)>0].explode('contentAsText').reset_index()
textDf = textDf[textDf.contentAsText.apply(len)>0].rename(columns = {'contentAsText':'content'})
textDf['page'] = textDf['content'].apply(lambda page_content : page_content[0])
textDf['content'] = textDf['content'].apply(lambda page_content : page_content[1])
textDf = textDf[textDf.content.apply(len)>0]
textDf.content = textDf.content.apply(removeNoisySection)
textDf = textDf[textDf.content.apply(len)>1000]
```
